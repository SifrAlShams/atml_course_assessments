```python
# Colab-ready FedAvg experiments: K sweep and sampling fraction sweep
# Paste this into a Colab cell and run. Assumes a GPU (cuda) is available.

# 0) Installs (torch usually present; safe to ensure torchvision)
!pip -q install torch torchvision --upgrade

# 1) Imports & seeding
import torch, torch.nn as nn, torch.nn.functional as F, torch.optim as optim
import torchvision, torchvision.transforms as transforms
from torch.utils.data import Subset, DataLoader, ConcatDataset
import numpy as np, random, copy, math, time
import matplotlib.pyplot as plt
from tqdm import trange

# Deterministic-ish seeds
SEED = 12345
torch.manual_seed(SEED)
np.random.seed(SEED)
random.seed(SEED)

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Device:", device)

# 2) Dataset: CIFAR-10 small subset for speed (change total_samples for bigger runs)
transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize((0.4914,0.4822,0.4465), (0.247,0.243,0.261))
])
full_train = torchvision.datasets.CIFAR10(root='./data', train=True, download=True, transform=transform)
test_set = torchvision.datasets.CIFAR10(root='./data', train=False, download=True, transform=transform)

# Use a subset for fast iteration; ensure divisible by M
total_samples = 5000   # increase if you have more time (e.g., 10000, 20000)
M = 5                  # number of clients
assert total_samples % M == 0
per_client = total_samples // M

indices = list(range(len(full_train)))
random.shuffle(indices)
subset_indices = indices[:total_samples]
client_indices = [subset_indices[i*per_client:(i+1)*per_client] for i in range(M)]
client_datasets = [Subset(full_train, idxs) for idxs in client_indices]
client_loaders_fullbatch = [DataLoader(ds, batch_size=len(ds), shuffle=True) for ds in client_datasets]

# central combined subset loader (for evaluation/training loss if needed)
combined_subset = Subset(full_train, subset_indices)
central_loader = DataLoader(combined_subset, batch_size=256, shuffle=False)

# test loader
test_loader = DataLoader(test_set, batch_size=1000, shuffle=False)

print("Using CIFAR-10 subset:", total_samples, "samples,", "clients:", M, "per client:", per_client)

# 3) Simple 2-layer CNN
class SmallCNN(nn.Module):
    def __init__(self, num_classes=10):
        super().__init__()
        self.conv1 = nn.Conv2d(3, 32, kernel_size=3, padding=1)  # 32x32
        self.conv2 = nn.Conv2d(32, 64, kernel_size=3, padding=1) # 32x32
        self.pool = nn.MaxPool2d(2,2)                            # 16x16
        self.fc1 = nn.Linear(64*16*16, 128)
        self.fc2 = nn.Linear(128, num_classes)
    def forward(self,x):
        x = F.relu(self.conv1(x))
        x = F.relu(self.conv2(x))
        x = self.pool(x)
        x = x.view(x.size(0), -1)
        x = F.relu(self.fc1(x))
        x = self.fc2(x)
        return x

# 4) Helpers: copy params, l2 diff, evaluate
def get_param_tensors(model):
    return [p.data.clone() for p in model.parameters()]

def set_param_tensors(model, tensors):
    for p, t in zip(model.parameters(), tensors):
        p.data.copy_(t)

def add_param_tensors(target, src, alpha=1.0):
    with torch.no_grad():
        for t, s in zip(target.parameters(), src):
            t.data.add_(alpha * s.to(t.device))

def param_l2_norm_diff(paramsA, paramsB):
    s = 0.0
    for a,b in zip(paramsA, paramsB):
        diff = (a - b).view(-1)
        s += float((diff*diff).sum().cpu())
    return math.sqrt(s)

def evaluate(model, loader, device):
    model.eval()
    correct = 0
    total = 0
    running_loss = 0.0
    criterion = nn.CrossEntropyLoss(reduction='sum')
    with torch.no_grad():
        for xb, yb in loader:
            xb, yb = xb.to(device), yb.to(device)
            out = model(xb)
            running_loss += float(criterion(out, yb).item())
            preds = out.argmax(dim=1)
            correct += int((preds == yb).sum().item())
            total += yb.size(0)
    return running_loss / total, correct / total

# 5) FedAvg implementation (single function to run experiments)
def run_fedavg_experiment(K, frac, rounds, lr=0.01, local_bs=64, momentum=0.9, verbose=True):
    """
    K: local epochs per client
    frac: sampling fraction f in (0,1] (if 1.0 -> full participation)
    rounds: number of communication rounds
    Returns: dict with logs (global_acc, global_loss, drift_per_round, comm_cost_est)
    """
    # initialize global model and identical initial weights for reproducibility
    global_model = SmallCNN().to(device)
    init_params = get_param_tensors(global_model)

    # create per-client models and dataloaders (mini-batches)
    client_models = [SmallCNN().to(device) for _ in range(M)]
    for cm in client_models:
        set_param_tensors(cm, init_params)

    client_loaders = [DataLoader(ds, batch_size=local_bs, shuffle=True) for ds in client_datasets]

    # logging
    global_accs = []
    global_losses = []
    drift_vals = []   # average L2 distance between client final model and global model (before aggregation)
    comm_costs = []   # estimated bytes (model_size * n_clients) per round (rough)
    model_param_count = sum(p.numel() for p in global_model.parameters())

    for r in range(rounds):
        # sample clients
        n_clients = max(1, int(round(frac * M)))
        selected = random.sample(range(M), n_clients) if n_clients < M else list(range(M))

        # store global params before local updates
        global_params_before = get_param_tensors(global_model)

        client_params_after = []
        client_data_sizes = []

        # For each selected client: copy global params, run K local epochs with optimizer, return params
        for i in selected:
            # reset client's model to global
            set_param_tensors(client_models[i], global_params_before)

            optimizer = optim.SGD(client_models[i].parameters(), lr=lr, momentum=momentum)
            client_models[i].train()
            # local training for K epochs
            loader = client_loaders[i]
            for epoch in range(K):
                for xb, yb in loader:
                    xb, yb = xb.to(device), yb.to(device)
                    optimizer.zero_grad()
                    out = client_models[i](xb)
                    loss = F.cross_entropy(out, yb)
                    loss.backward()
                    optimizer.step()

            # record final params
            client_params_after.append(get_param_tensors(client_models[i]))
            client_data_sizes.append(len(client_datasets[i]))

        # compute drift: average L2 distance between each selected client's params and global before aggregation
        drift_sum = 0.0
        for params in client_params_after:
            drift_sum += param_l2_norm_diff(params, global_params_before)
        avg_drift = drift_sum / len(client_params_after)
        drift_vals.append(avg_drift)

        # Aggregation: weighted average by data size (Ni / sum Ni_selected)
        total_selected_data = sum(client_data_sizes)
        new_global_params = [torch.zeros_like(p) for p in global_params_before]
        for params, Ni in zip(client_params_after, client_data_sizes):
            weight = Ni / total_selected_data
            for idx, p in enumerate(params):
                new_global_params[idx] += weight * p.to(new_global_params[idx].device)

        # update global model
        set_param_tensors(global_model, new_global_params)

        # logging: evaluate after aggregation
        g_loss, g_acc = evaluate(global_model, test_loader, device)
        global_losses.append(g_loss)
        global_accs.append(g_acc)

        # communication cost estimate: model_param_count * 4 bytes * n_clients (float32) -> bytes per round
        comm_bytes = model_param_count * 4 * len(selected)
        comm_costs.append(comm_bytes)

        if verbose:
            print(f"Round {r+1}/{rounds} | selected {len(selected)}/{M} clients | drift {avg_drift:.4e} | test acc {g_acc:.4f}")

    return {
        'global_accs': global_accs,
        'global_losses': global_losses,
        'drift': drift_vals,
        'comm_costs': comm_costs,
        'model_param_count': model_param_count
    }

# 6) Experiment 1: varying K with full participation (frac=1.0)
rounds = 30
K_list = [1, 5, 10, 20]
results_K = {}

print("\n=== Experiment 1: Varying K (full participation) ===")
for K in K_list:
    print(f"\nRunning FedAvg with K={K}, full participation (frac=1.0)")
    res = run_fedavg_experiment(K=K, frac=1.0, rounds=rounds, lr=0.01, local_bs=64, momentum=0.9, verbose=True)
    results_K[K] = res

# 7) Plot Experiment 1: accuracy vs rounds for different K
plt.figure(figsize=(8,5))
for K, res in results_K.items():
    plt.plot(range(1, rounds+1), res['global_accs'], label=f"K={K}")
plt.xlabel("Communication round")
plt.ylabel("Test accuracy")
plt.title("FedAvg: varying K (full participation)")
plt.legend()
plt.grid(True)
plt.show()

# 8) Also plot drift vs rounds (avg drift per round) for K variants
plt.figure(figsize=(8,5))
for K, res in results_K.items():
    plt.plot(range(1, rounds+1), res['drift'], label=f"K={K}")
plt.xlabel("Communication round")
plt.ylabel("Avg client drift (L2)")
plt.title("Client drift per round: varying K (full participation)")
plt.legend()
plt.grid(True)
plt.show()

# 9) Experiment 2: vary sampling fraction f while keeping K small (e.g., K=5)
K_fixed = 5
f_list = [1.0, 0.5, 0.2]
results_f = {}

print("\n=== Experiment 2: Varying sampling fraction f (K fixed = {}) ===".format(K_fixed))
for f in f_list:
    print(f"\nRunning FedAvg with K={K_fixed}, frac={f}")
    res = run_fedavg_experiment(K=K_fixed, frac=f, rounds=rounds, lr=0.01, local_bs=64, momentum=0.9, verbose=True)
    results_f[f] = res

# 10) Plot Experiment 2: accuracy vs rounds for different fractions
plt.figure(figsize=(8,5))
for f, res in results_f.items():
    plt.plot(range(1, rounds+1), res['global_accs'], label=f"f={f}")
plt.xlabel("Communication round")
plt.ylabel("Test accuracy")
plt.title(f"FedAvg: varying sampling fraction (K={K_fixed})")
plt.legend()
plt.grid(True)
plt.show()

# 11) Plot drift vs rounds for sampling fractions
plt.figure(figsize=(8,5))
for f, res in results_f.items():
    plt.plot(range(1, rounds+1), res['drift'], label=f"f={f}")
plt.xlabel("Communication round")
plt.ylabel("Avg client drift (L2)")
plt.title(f"Client drift per round: varying fraction (K={K_fixed})")
plt.legend()
plt.grid(True)
plt.show()

# 12) Summary tables: final accuracies and total communication bytes (rough)
print("\n--- Summary: final test accuracies and total communication (approx bytes) ---")
for K, res in results_K.items():
    total_comm = sum(res['comm_costs'])
    print(f"K={K}: final acc={res['global_accs'][-1]:.4f}, avg drift (last 5 rounds)={np.mean(res['drift'][-5:]):.4e}, total_comm_bytes={total_comm:,}")

for f, res in results_f.items():
    total_comm = sum(res['comm_costs'])
    print(f"f={f} (K={K_fixed}): final acc={res['global_accs'][-1]:.4f}, avg drift (last 5 rounds)={np.mean(res['drift'][-5:]):.4e}, total_comm_bytes={total_comm:,}")

print("\nDone. Plots show how K and sampling fraction affect accuracy and drift.")
```
